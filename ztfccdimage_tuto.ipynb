```python
from ztfccdimage import ccdimage
```

```python
# Initialization of image object
imgfile = '/Users/philippe/Data/ZTF/raw/2020/0301/124711/ztf_20200301124711_001508_zr_c06_o.fits.fz'
img = ccdimage(imgfile)
```

```python
# To get raw CCD image (by default corrected from overscan and bias)
ccd_raw = img.get_ccd_raw()
```

```python
# To get global CCD flat-filed for the image: compute it if not existing
ccd_flat = img.get_ccd_flat()
```

```python
# To get CCD bias for thi image 
ccd_bias = img.get_ccd_bias()
```

```python
# To get CCD image from ztf pipeline
ccd_ztf_sci = img.get_ztf_ccd_sci()
```

```python
# To get CCD image with global CCD flat-field: home-made starting from raw image
ccd_my_sci = img.get_my_ccd_sci()
```
